```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
import seaborn as sns

#Only get data analyst jobs in the US
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_DA_US_Pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_DA_US_Pivot.loc['total'] = df_DA_US_Pivot.sum()
df_DA_US_Pivot
```

```python
df_DA_US_Pivot[df_DA_US_Pivot.loc['total'].sort_values(ascending=False).index]
```

```python
df_DA_US_Pivot = df_DA_US_Pivot[df_DA_US_Pivot.loc['total'].sort_values(ascending=False).index]
```

```python
df_DA_US_Pivot = df_DA_US_Pivot.drop('total')
```

```python
df_DA_total = df_DA_US.groupby('job_posted_month_no').size()
```

```python
df_DA_percent = df_DA_US_Pivot.div(df_DA_total/100, axis=0)

df_DA_percent
```

```python
#changes month number to month name
df_DA_percent.index = df_DA_percent.index.map({1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'})
```

```python
df_DA_percent
```

```python
df_Plot = df_DA_percent.iloc[:, :5]

sns.lineplot(data=df_Plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analyst Jobs in the US')
plt.xlabel('2023')
plt.ylabel('likelihood in job posting')
plt.legend().remove()


from matplotlib.ticker import PercentFormatter
ax= plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

for i in range(5):
    plt.text(11.2, df_Plot.iloc[-1, i], df_Plot.columns[i], fontsize=9, color='black')


plt.show()
```

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US_software_pivot = pd.read_csv('https://lukeb.co/software_csv', index_col='job_posted_month')
```

```python
df_US_software_pivot
```

```python
df_US_pivot = pd.read_pickle('df_US_pivot.pkl')
```

```python
#let run the merge function to merge the two dataframes

df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
```

```python
df_US_merge = df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
```

```python
top_5 = (
    df_US_merge
    .sum().
    sort_values(ascending=False).
    head(5).
    index
    .to_list()
 )

top_5
```

```python
df_US_merge[top_5].plot(kind='line')
plt.title('monthly job postings for top tech jobs in the US')
plt.ylabel('job counts')
plt.xlabel('2023')
plt.ylim(0, 20000)
plt.legend()
plt.show()
```

```python
#loading the libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading the dataset, puting in a dataframe and checking first 5 rows

dataset = load_dataset("lukebarousse/data_jobs")
df =dataset['train'].to_pandas()
df.head()

#cleaning the data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States']
```

```python
df_US = df[df['job_country'] == 'United States'].copy()
```

```python
df_US
```

```python
#formating date 

df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')
```

```python
df_US
```

```python
df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_US_pivot.reset_index(inplace=True)
```

```python
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month
```

```python
df_US_pivot
```

```python
#order by month

df_US_pivot.sort_values('month_no', inplace=True)
df_US_pivot.set_index('job_posted_month', inplace=True)
```

```python
#month ordered

df_US_pivot
```

```python
df_US_pivot.sort_values('month_no', inplace=True)
df_US_pivot.drop(columns='month_no', inplace=True)
```

```python
df_US_pivot
```

```python
#plt line chart

df_US_pivot.plot(kind='line')
plt.show()
```

```python
#top 3 job titles

top_3 = df_US['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_US_pivot[top_3].plot(kind='line')
plt.title('Monthly job postings for top data job in the US')
plt.ylabel('job counts')
plt.xlabel('2023')  
plt.legend()
plt.show()
```

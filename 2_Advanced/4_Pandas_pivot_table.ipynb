```python
#loading the libraries

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading the dataset, puting in a dataframe and checking first 5 rows

dataset = load_dataset("lukebarousse/data_jobs")
df =dataset['train'].to_pandas()
df.head()

#cleaning the data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

##**Counts of jobs titles**


```python
#using the pivot table to get the count of job titles 

df.pivot_table(index='job_title_short', aggfunc='size')
```

```python
# get the same result using groupby

df.groupby('job_title_short').size()
```

**median salary by job**

```python
df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
```

```python
#groupby method

df.groupby('job_title_short')['salary_year_avg'].median()
```

```python
df.pivot_table(values='salary_year_avg', index='job_title_short',columns='job_country',aggfunc='median')
```

```python
# changing inde xand columns

df.pivot_table(values='salary_year_avg', index='job_country',columns='job_title_short', aggfunc='median')
```

```python
#finding the top 6 countries 

top_6_countries =df['job_country'].value_counts().head(6).index
df_job_country_salary = df.pivot_table(
    values='salary_year_avg', 
    index='job_country',
    columns='job_title_short', 
    aggfunc='median'
    )

df_job_country_salary = df_job_country_salary.loc[top_6_countries]

job_title =[ 'Data Analyst', 'Data Engineer', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_title]

df_job_country_salary

```

```python
#plotting the data
df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary($USD)')
plt.xlabel('')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()
```
